#Topic 3: Hypothesis Testing

- a peek into https://www.statsmodels.org/stable/index.html 

```python
#don't type this

import statsmodels.api as sm
sleep = sm.datasets.get_rdataset("sleep").data
```

```python
#don't type this

sleep
```

```python
#don't type this

import pandas as pd

gp1 = sleep[sleep.group==1].extra
gp2 = sleep[sleep.group==2].extra

c = pd.DataFrame({'gp 1':gp1.values,'gp 2':gp2.values})
c.plot(kind='box')
```

```python
#don't type this

import scipy

result = scipy.stats.ttest_ind(gp1,gp2)
result.pvalue
```

```python
#don't type this

a1 = gp1.reset_index().extra
b1 = gp2.reset_index().extra
result = scipy.stats.ttest_1samp(a1-b1, 0)
result.pvalue
```

##Topic 4a: Linear Regression Model

---




```python
#don't type this

import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 20)
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)
data = pd.DataFrame({'x': x, 'y': y})
```

```python
#don't type this

from statsmodels.formula.api import ols
model = ols("y ~ x", data).fit()
```

```python
#don't type this

model.summary()
```

```python
#don't type this

a = model.params[0]
b= model.params[1]

yhat = a + b*x

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x,y,'or')
plt.plot(x,yhat,'b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#type this

import pandas as pd
mtcars = pd.read_csv('https://www.alvinang.sg/s/mtcars.csv')
```

```python
#type this

model = ols("mpg ~ wt", mtcars).fit()
```

```python
#type this

a = model.params[0]
b= model.params[1]

yhat = a + b*mtcars.wt
mtcars.plot.scatter('wt','mpg')
plt.plot(mtcars.wt,yhat,'r')
```
